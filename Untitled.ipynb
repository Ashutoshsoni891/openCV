```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image = cv2.imread('test5.jpeg')
grey_image = cv2.cvtColor(image , cv2.COLOR_BGR2RGB)
plt.figure()
plt.imshow(grey_image,cmap='gray')
```

```python
face_cascasde = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
face_cascasde
```

```python
face_detect = face_cascasde.detectMultiScale(grey_image , scaleFactor = 1.1,minNeighbors =1)
len(face_detect)
```

```python
for a,b,c,d in face_detect:
    cv2.rectangle(image , (a,b) , (a+c,b+d) , (0,0,255) , 1)
```

```python
plt.figure()
plt.imshow(cv2.cvtColor(image , cv2.COLOR_RGB2BGR))
```
